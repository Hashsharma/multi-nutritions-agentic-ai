```python

sq = [1, 2, 3, 4, 5]
target = 9
def binary_search(arr):

    left, right = 0, len(arr)-1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return arr[mid]
        elif arr[mid] < target:
            right = mid - 1
        elif arr[mid] > target:
            left = mid + 1
    
    return -1
    
binary_search(sq)
```

```python
s = "abacabcbb"
lookup = set()
left = 0
max_length = 0

for right in range(len(s)):
    while s[right] in lookup:
        lookup.remove(s[left])
        left += 1

    lookup.add(s[right])
    max_length = max(max_length, right - left + 1)

print(max_length)
```

```python
# a = '(', ')', '{', '}', '[', ']'

def is_valid(s: str) -> bool:
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}

    for ch in s:
        if ch in pairs:
            top = stack.pop() if stack else '#'
            if pairs[ch] != top:
                return False
        else:
            stack.append(ch)

    return not stack

print(is_valid("({)[]{}}}"))   # True
```
